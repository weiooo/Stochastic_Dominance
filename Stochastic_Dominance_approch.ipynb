## Define function

```python
def surface_right(targetx, targety, x1, y1, x2, y2):
    m = (y2 - y1) / (x2 - x1)
    b = y2 - m * x2
    if (m * targetx + b) >= targety:
        return True
    else:
        return False
```

```python
def examine_dominate(x1, y1, x2, y2):
    result = []
    if (x2[0] >= x1[0]):
        result.append(1)
        for i in range(1,len(x2)):
            index = []
            for element in x1:
                if element > x2[i]:
                    index.append(np.where(x1 == element)[0][0])
                    break
            if not index: # index == []
                index.append(len(x1)-1) # then we need to check x1[19], x[18] line

            LR = surface_right(x2[i], y2[i], x1[index[0]-1], y1[index[0]-1], x1[index[0]], y1[index[0]])
            if LR == True:
                result.append(1)
            else:
                result.append(0)
    else:
        result.append(0) # x2 will never dominate x1
    #print(result)
    
    if (0 in result): 
        #print("x2 does not dominate x1")
        return 0
    else:
        #print("x2 dominates x1 !!")
        return 1
```

## Analyse the data cross group and month

### Part I: January Return

#### (1) Compare JANUARY returns of 10 portfolios

```python
compare_month = 1 # JAN
specific_group = [0,3,6,9]

for i in range(10): #specific_group: 
    
    x = np.sort(df4.iloc[i][compare_month])/100
    y = np.arange(1, len(x)+1) / len(x)
    
    plt.xlabel('JANUARY PORTFOLIO RETURNS')
    plt.ylabel('CUMULATIVE PROBABILITY')
    #plt.plot(x, y, label = 'S'+str(i+1))
    plt.step(x, y, label='S'+str(i+1))
    
    plt.legend(loc='lower right', shadow=True)

#plt.title('FSD')
plt.show()
```

#### (2) Compare JANUARY returns between any two portfolios

```python
group = ['S'+str(i) for i in range(1,11)]
df = pd.DataFrame({'group':group})
for g in group:
    df[g] = '-'

df.set_index('group', inplace=True)

for i in range(10):
    all_group = [x for x in range(10)]
    all_group.pop(i)
    for j in all_group:
        
        x1 = np.sort(df4.iloc[j][compare_month])/100 # Si JAN
        x2 = np.sort(df4.iloc[i][compare_month])/100 # S jJAN

        # FSD
        y1_1 = np.arange(1, len(x1)+1) / len(x1)
        y1_2 = np.arange(1, len(x2)+1) / len(x2)

        if (x1[-1] < x2[-1]):
            x1 = np.append(x1, x2[-1])
            y1_1 = np.append(y1_1, 1) # FSD
            # SSD
            y2_1 = np.cumsum([0] + [(x1[i] - x1[i-1]) * y1_1[i-1] for i in range(1,len(x1))])
            y2_2 = np.cumsum([0] + [(x2[i] - x2[i-1]) * y1_2[i-1] for i in range(1,len(x2))])
            # TSD
            y3_1 = np.cumsum([0] + [(x1[i] - x1[i-1]) * (y2_1[i] + y2_1[i-1])/2 for i in range(1,len(x1))])
            y3_2 = np.cumsum([0] + [(x2[i] - x2[i-1]) * (y2_2[i] + y2_2[i-1])/2 for i in range(1,len(x2))])

        else:
            x2 = np.append(x2, x1[-1])
            y1_2 = np.append(y1_2, 1) # FSD
            # SSD
            y2_1 = np.cumsum([0] + [(x1[i] - x1[i-1]) * y1_1[i-1] for i in range(1,len(x1))])
            y2_2 = np.cumsum([0] + [(x2[i] - x2[i-1]) * y1_2[i-1] for i in range(1,len(x2))])
            # TSD
            y3_1 = np.cumsum([0] + [(x1[i] - x1[i-1]) * (y2_1[i] + y2_1[i-1])/2 for i in range(1,len(x1))])
            y3_2 = np.cumsum([0] + [(x2[i] - x2[i-1]) * (y2_2[i] + y2_2[i-1])/2 for i in range(1,len(x2))])

        
        fsd = examine_dominate(x1, y1_1, x2, y1_2)
        if fsd == 1:
            #print('Group',i+1,'JANUARY dominate Group',j+1, 'by FSD')
            result = 'FSD'
        else:
            ssd = examine_dominate(x1, y2_1, x2, y2_2)
            if ssd == 1:
                #print('Group',i+1,'JANUARY dominate Group',j+1, 'by SSD')
                result = 'SSD'
            else:
                tsd = examine_dominate(x1, y3_1, x2, y3_2)
                if tsd == 1:
                    #print('Group',i+1,'JANUARY dominate Group',j+1, 'by TSD')
                    result = 'TSD'
                else:
                    #print('Group',i+1,'JANUARY does not dominate Group',j+1)
                    result = ' '
        
        df.iloc[i, j] = result 
        #print('Group',i+1,'JANUARY dominate Group',j+1,'by',result)
        
df
```
